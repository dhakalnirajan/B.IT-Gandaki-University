```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("C:/Saugat-Programs/Semester VII/Data Science/Data Visualization/netflix_titles.csv")

df = pd.DataFrame(data, columns=["duration","release_year","type"])

sns.histplot(df["duration"],kde=True)
plt.title("Movie Duration")
plt.show()
```

```python
sns.lineplot(df["release_year"])
# plt.plot(df["release_year"])
plt.title("Release Year")
plt.show()
```

```python
type_counts = df["type"].value_counts().reset_index()
type_counts.columns = ["type", "count"]

# Plot using barplot
sns.barplot(x="type", y="count", data=type_counts)
plt.title("Show Types")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
```

```python
year_counts = df['release_year'].value_counts()
highest_year = year_counts.idxmax()
highest_count = year_counts.max()

print("Year with highest releases:", highest_year)
print("Number of releases:", highest_count)

year_counts.sort_index().plot(kind='bar')
plt.title("Number of Releases per Year")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.show()
```
