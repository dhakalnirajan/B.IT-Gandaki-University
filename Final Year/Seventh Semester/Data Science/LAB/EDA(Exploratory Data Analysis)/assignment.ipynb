```python
import pandas as pd
import numpy as np

data = pd.read_csv("C:/Saugat-Programs/Semester VII/Data Science/EDA(Exploratory Data Analysis)/Students Performance.csv")

df = pd.DataFrame(data,columns=["Math_Score"])

mean = df["Math_Score"].mean()
median = df["Math_Score"].median()
mode = df["Math_Score"].mode()[0]
variance = df["Math_Score"].var()
std_dev = df["Math_Score"].std()
IQR = np.percentile(df["Math_Score"],75) - np.percentile(df["Math_Score"],25)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Inter Quartile Range(IQR): {IQR}")
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv("C:/Saugat-Programs/Semester VII/Data Science/EDA(Exploratory Data Analysis)/Students Performance.csv")

df = pd.DataFrame(data,columns=["Math_Score","Reading_Score","Writing_Score","Placement_Score"])

mean = df.mean()
median = df.median()
mode = df.mode().iloc[0]
variance = df.var()
std_dev = df.std()
IQR = np.percentile(df,75) - np.percentile(df,25)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Inter Quartile Range(IQR): {IQR}")
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv("C:/Saugat-Programs/Semester VII/Data Science/EDA(Exploratory Data Analysis)/Students Performance.csv")

df = pd.concat([data["Math_Score"],data["Reading_Score"],data["Writing_Score"],data["Placement_Score"]])

mean = df.mean()
median = df.median()
mode = df.mode().iloc[0]
variance = df.var()
std_dev = df.std()
IQR = np.percentile(df,75) - np.percentile(df,25)

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Mode: {mode}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Inter Quartile Range(IQR): {IQR}")
```

```python
# More Used then concat
data['Total_Score'] = data[['Math_Score', 'Reading_Score', 'Writing_Score', 'Placement_Score']].mean(axis=1)
mean_total = data['Total_Score'].mean()
median_total = data['Total_Score'].median()
mode_total = data['Total_Score'].mode()[0]
variance_total = data['Total_Score'].var()
std_dev_total = data['Total_Score'].std()
print(f"Mean Total Score: {mean_total}")
print(f"Median Total Score: {median_total}")
print(f"Mode Total Score: {mode_total}")
print(f"Variance Total Score: {variance_total}")
print(f"Standard Deviation Total Score: {std_dev_total}")
```
