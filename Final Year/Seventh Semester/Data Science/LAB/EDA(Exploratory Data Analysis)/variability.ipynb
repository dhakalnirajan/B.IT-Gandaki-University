```python
import pandas as pd
import numpy as np

#data
data = [24,45,67,89,23,56,45,67,46,38,45]

#Create DataFrame
df = pd.DataFrame(data,columns=["Values"])

#calculate range, variance, standard Deviation
range_value = df["Values"].max()-df["Values"].min()
variance = df["Values"].var()
std_dev = df["Values"].std()
IQR = np.percentile(df["Values"],75) - np.percentile(df["Values"],25)
print(f"Range:{range_value}")
print(f"Variance:{variance}")
print(f"Standard Deviation:{std_dev}")
print(f"IQR:{IQR}")
```
