```python
# Task 1:
# Is there any significance difference between the marks of male and female students
import pandas as pd
from scipy.stats import ttest_ind

# Load the dataset (update path if needed)
df = pd.read_csv("C:/Saugat-Programs/Semester VII/Data Science/Statistical Analysis/StudentsPerformance.csv")

# Separate math scores by gender
male_math = df[df["gender"] == "male"]["math score"]
female_math = df[df["gender"] == "female"]["math score"]

# Perform Welch’s t-test (equal_var=False)
t_statistic, p_value = ttest_ind(
    male_math,
    female_math,
    equal_var=False
)

# Significance level
alpha = 0.05

# Print results
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Decision
if p_value <= alpha:
    print("Result: Significant difference in math scores between male and female students.")
else:
    print("Result: No significant difference in math scores between male and female students.")
```

```python
# Task 2:
# Covid-19 patient pre-condition dataset (kaggle):
# is the mean age significantly difference between the recovered and decreased patient
import pandas as pd
from scipy.stats import ttest_ind

# Load dataset safely
df = pd.read_csv(
    "C:/Saugat-Programs/Semester VII/Data Science/Statistical Analysis/patient.csv",
    low_memory=False
)

# Separate groups using OUTCOME column
recovered_age = df[df["outcome"] == 1]["age"]
deceased_age  = df[df["outcome"] == 2]["age"]

# Drop missing values
recovered_age = recovered_age.dropna()
deceased_age = deceased_age.dropna()

# Check sample sizes
print("Recovered sample size:", len(recovered_age))
print("Deceased sample size:", len(deceased_age))

# Welch’s t-test
t_stat, p_value = ttest_ind(
    recovered_age,
    deceased_age,
    equal_var=False
)

# Results
print(f"Recovered patients mean age: {recovered_age.mean():.2f}")
print(f"Deceased patients mean age: {deceased_age.mean():.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.6f}")

# Conclusion
alpha = 0.05
if p_value <= alpha:
    print("Conclusion: Mean age is significantly different between recovered and deceased patients.")
else:
    print("Conclusion: No significant difference in mean age.")

```
