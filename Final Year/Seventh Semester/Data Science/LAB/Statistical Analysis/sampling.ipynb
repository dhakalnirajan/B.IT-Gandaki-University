```python
# Cluster Sampling
import pandas as pd
import numpy as np

data = {
    "School": ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L", "M", "N", "O"],
    "Marks": [85, 78, 92, 66, 74, 88, 90, 81, 77, 69, 95, 83, 72, 80, 87]
}

df = pd.DataFrame(data)

# Step 1: Randomly selects two clusters(School)
selected_schools = np.random.choice(df["School"].unique(), size=2, replace=False)

# Step 2: Select all students from selected cluster
cluster_sample = df[df["School"].isin(selected_schools)]

# Step 3: Calculate average marks
average_marks = cluster_sample["Marks"].mean()

# Output
print("Selected Schools:", selected_schools)
print("Average Marks", average_marks)
```
