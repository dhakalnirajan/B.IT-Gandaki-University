```python
# Day 1
from scipy.stats import ttest_1samp

data = [49.5,50.3,50.3,50.9,50.0,49.4,49.2,49.8,50.6,49.8]

t_stat, p_value = ttest_1samp(data,50)
alpha = 0.05

print("T-statistics:", t_stat)
print("P-Value:", p_value)
if p_value < alpha:
    print("Reject the null hypothesis. The mean is significantly different from 50.")
else:
    print("Fail to reject the null hypothesis. No significant difference from 50.")
```

```python
# Day 2
from scipy.stats import ttest_ind

# Sample Data
group_a = [80,85,87,90,75,79,79]
group_b = [74,75,78,74,68,78,73]

# Perform two-sample t-test with Welch's correction
t_stat, p_value = ttest_ind(group_a,group_b,equal_var=False)

# Significance_level
alpha = 0.05

# Results
print("T-statistics:", t_stat)
print("P-Value:", p_value)
if p_value < alpha:
    print("Reject the null hypothesis. The mean is significantly different.")
else:
    print("Fail to reject the null hypothesis. No significant different.")
```
