```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'Age': [35,np.nan,40,45,np.nan],
    'Income': [np.nan,60000,np.nan,70000,80000]
})

#removing row
filtered_data = data.dropna()
print(filtered_data)
```
