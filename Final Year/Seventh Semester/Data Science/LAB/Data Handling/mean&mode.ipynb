```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'Age': [35,np.nan,40,45,np.nan],
    'Income': [50000,60000,np.nan,70000,80000],
    'Gender': ['F','M','F',np.nan,'M']
})

#filling missing numerical data with mean
data['Age']=data['Age'].fillna(data['Age'].mean()).astype(int)
data['Income']=data['Income'].fillna(data['Income'].mean()).astype(int)

#filling missing non-numerical data with mode
data['Gender']=data['Gender'].fillna(data['Gender'].mode()[0])
print(data)
```
