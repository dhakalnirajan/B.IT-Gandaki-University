```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'Gender': ['Boy','Girl','Boy','Girl',np.nan,'Girl'],
    'Height': [150,145,np.nan,152,155,np.nan]
})

data['Gender']=data['Gender'].fillna(data['Gender'].mode()[0])
data['Height']=data['Height'].fillna(data['Height'].mean()).astype(int)
print(data)
```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('C:/Saugat-Programs/Semester VII/Data Science/Data Handling/train.csv')

original_age = data['Age'].copy()
original_name = data['Name'].copy()

data['PassengerId'] = data['PassengerId'].fillna(data['PassengerId'].mean()).astype(int)
data['Survived'] = data['Survived'].fillna(data['Survived'].mean()).astype(int)
data['Age'] = data['Age'].fillna(data['Age'].mean()).astype(int)
data['Name'] = data['Name'].fillna(data['Name'].mode()[0])
data['Sex'] = data['Sex'].fillna(data['Sex'].mode()[0])
data['SibSp'] = data['SibSp'].fillna(data['SibSp'].mean()).astype(int)
data['Fare'] = data['Fare'].fillna(data['Fare'].mean()).astype(int)
data['Embarked'] = data['Embarked'].fillna(data['Embarked'].mode()[0])

age_comparison = pd.concat(
    [original_age, data['Age']],
    axis=1,
    keys=['Age_Before', 'Age_After']
)

name_comparison = pd.concat(
    [original_name, data['Name']],
    axis=1,
    keys=['Name_Before','Name_After']
)

print(age_comparison.head(15))
print(name_comparison.head(15))
```
