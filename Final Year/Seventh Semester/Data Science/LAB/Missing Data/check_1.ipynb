```python
import pandas as pd
import numpy as np

data = pd.DataFrame({
    'Age': [10,np.nan,12,np.nan,15],
    'Income': [10000,20000,np.nan,np.nan,50000]
})

#work of both isna() & isnull() is same. available = false, not-available = true
print("using isna():")
print(data.isna())

print("\nusing isnull()")
print(data.isnull())
```
