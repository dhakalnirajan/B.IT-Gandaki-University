```python
import pandas as pd

# Dataset A
df1 = pd.DataFrame({
    'id':[1,2,3,4],
    'Name':['Naruto','Luffy','Hinata','Hancock'],
    'Age':[18,17,16,16]
})

df2=pd.DataFrame({
    'customer_id':[1,2,3,4],
    'City':['Konoha','Red Line','Hidden Leaf','Vegapunk'],
})

merged_df=pd.merge(df1,df2,left_on='id', right_on='customer_id',how='inner')
print(merged_df)
```
