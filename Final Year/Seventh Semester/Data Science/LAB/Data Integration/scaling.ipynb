```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Original dataset (use LISTS, not SETS)
data = {
    'Age': [25, 30, 35, 40, 45],
    'Salary': [25000, 35000, 45000, 55000, 65000]
}

# Create DataFrame
df = pd.DataFrame(data)

# Apply Min-Max Scaling
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(df)

# Convert back to DataFrame
normalized_df = pd.DataFrame(
    normalized_data,
    columns=df.columns
)

# Display results
print("Original Data:")
print(df)

print("\nNormalized Data (Min-Max Scaling):")
print(normalized_df)
```
