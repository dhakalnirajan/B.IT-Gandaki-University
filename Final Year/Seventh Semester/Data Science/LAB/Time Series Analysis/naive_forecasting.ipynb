```python
import pandas as pd

# Load CSV data
data = pd.read_csv("C:/Saugat-Programs/Semester VII/Data Science/LAB/Time Series Analysis/shampoo_sales.csv")

# Create Naive Forecast column
data["Naive_Forecast"] = data["Sales"].shift(1)

# Display result
print(data)

# Forecast for next month (after last row)
next_month_forecast = data["Sales"].iloc[-1]
print("\nNaive Forecast for next month:", next_month_forecast)
```

```python
import matplotlib.pyplot as plt

plt.plot(data["Sales"], label="Actual Sales")
plt.plot(data["Naive_Forecast"], label="Naive Forecast")
plt.legend()
plt.show()
```
