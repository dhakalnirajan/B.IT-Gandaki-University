# Question 3: Create a dataset as below and identify the duplicated row with reference to the attribute “CustomerID” and remove the duplicated row.

````python
import pandas as pd
import numpy as np
````

````python
# Create the dataset
data = {
    'CustomerID': ['C001', 'C002', 'C003', 'C004', 'C004', 'C005', 'C006', 'C007', 'C008', 'C008'],
    'Name': ['John', 'Alice', 'BOB', 'BOBY', 'Eve', 'eve', 'Steve', 'Ramu', 'mary', 'Bob'],
    'Age': [25, 34, 17, 29, 29, 120, -5, np.nan, 220, 30],
    'JoinDate': ['12/1/2024', '11/15/2023', '6/1/2022', '6/1/2022', '12/5/2024', 'invalid_date', np.nan, '1/1/2024', '3/5/2023', '3/5/2023'],
    'MonthlyCharges': [29.85, 56.95, 4000, 75.5, 75.5, 45.99, 60, 49.99, -30, 55],
    'Churn': ['No', 'Yes', 'No', 'No', 'No', 'Yes', 'No', np.nan, 'Yes', 'No']
}

df = pd.DataFrame(data)

print("Original DataFrame:")
display(df)
````

````python
# Identify duplicated rows with reference to 'CustomerID'
duplicates = df[df.duplicated('CustomerID', keep=False)]
print("\nRows with duplicated CustomerID:\n")
display(duplicates)
````

````python
# Remove duplicated rows (keeping the first occurrence)
df_cleaned = df.drop_duplicates(subset='CustomerID', keep='first')

print("\nDataFrame after removing duplicates (based on CustomerID):\n")
display(df_cleaned)
````

 # Question 4: Create a dataset as below:

```
data = { "Feature1": [10, 20, 30, 40, 50],
"Feature2": [100, 200, 300, 400, 500],
"Feature3": [5, 15, 25, 35, 45]
}
```


1.   Perform Min-Max Normalization
2.   Perform Z-score Normalization
3.   Perform Decimal Scale Normalization

````python
import pandas as pd
import numpy as np

# Create the dataset
data = {
    "Feature1": [10, 20, 30, 40, 50],
    "Feature2": [100, 200, 300, 400, 500],
    "Feature3": [5, 15, 25, 35, 45]
}
df_norm = pd.DataFrame(data)
print("Original Dataset:")
display(df_norm)
````

````python
# 1. Min-Max Normalization: (x - min) / (max - min)
df_min_max = (df_norm - df_norm.min()) / (df_norm.max() - df_norm.min())
print("1. Min-Max Normalized Dataset:")
display(df_min_max)
````

````python
# 2. Z-score Normalization: (x - mean) / std
df_zscore = (df_norm - df_norm.mean()) / df_norm.std()
print("2. Z-score Normalized Dataset:")
display(df_zscore)
````

````python
# 3. Decimal Scale Normalization: x / 10^j where 10^j is the smallest power of 10 such that max(|x|)/10^j < 1
def decimal_scaling(df):
    df_scaled = df.copy()
    for col in df_scaled.columns:
        p = df_scaled[col].abs().max()
        j = np.ceil(np.log10(p))
        df_scaled[col] = df_scaled[col] / (10**j)
    return df_scaled

df_decimal = decimal_scaling(df_norm)
print("3. Decimal Scale Normalized Dataset:")
display(df_decimal)
````

# Question 5: Write python code to implement KNN classifier (with K =3) for predicting the new point `BMI= 43.6` and `Age= 40` with reference to below dataset.

| BMI | Age | Sugar |
|--- |--- |--- |
|33.6|50|1|
|26.6|30|0|
|23.4|40|0|
|43.1|67|0|
|35.3|23|1|
|35.9|67|1|
|36.7|45|1|
|25.7|46|0|
|23.3|29|0|
|31|56|1|

````python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier

# Create the dataset
data = {
    'BMI': [33.6, 26.6, 23.4, 43.1, 35.3, 35.9, 36.7, 25.7, 23.3, 31],
    'Age': [50, 30, 40, 67, 23, 67, 45, 46, 29, 56],
    'Sugar': [1, 0, 0, 0, 1, 1, 1, 0, 0, 1]
}

df_knn = pd.DataFrame(data)
display(df_knn)
````

````python
# Features and Target
X = df_knn[['BMI', 'Age']]
y = df_knn['Sugar']

# Initialize KNN with K=3
knn = KNeighborsClassifier(n_neighbors=3)

# Fit the model
knn.fit(X, y)

# Define the new point
new_point = pd.DataFrame([[43.6, 40]], columns=['BMI', 'Age'])

# Predict
prediction = knn.predict(new_point)

print(f"Prediction for BMI=43.6, Age=40: Sugar = {prediction[0]}")
````

# Question 6: You are provided with the following dataset:

Staff Age Weight Color
101 5 20 Brown
102 3 8 Black
103 2 0.5 Yellow
104 4 18 White
105 7 9 Brown

Perform the following tasks:

a) Apply one-hot encoding to Color columns.

b) Combine the encoded features with the rest of the dataset.

c) Display the resulting DataFrame.

````python
import pandas as pd
import matplotlib.pyplot as plt
# Create the dataset
data = {
    'Staff': [101, 102, 103, 104, 105],
    'Age': [5, 3, 2, 4, 7],
    'Weight': [20, 8, 0.5, 18, 9],
    'Color': ['Brown', 'Black', 'Yellow', 'White', 'Brown']
}

df_staff = pd.DataFrame(data)
print("Original DataFrame:")
display(df_staff)
````

````python
# a) Apply one-hot encoding to Color columns
color_encoded = pd.get_dummies(df_staff['Color'], prefix='Color')
display(color_encoded)
````

````python
# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 2))

# Hide axes
ax.xaxis.set_visible(False)
ax.yaxis.set_visible(False)
ax.set_frame_on(False)

# Create the table from color_encoded dataframe
table = ax.table(cellText=color_encoded.values, colLabels=color_encoded.columns, cellLoc='center', loc='center')

# Style the table
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)

plt.title('Color Encoded DataFrame Visualization')
plt.show()
````

````python
# b) Combine the encoded features with the rest of the dataset
df_final = pd.concat([df_staff.drop('Color', axis=1), color_encoded], axis=1)

# c) Display the resulting DataFrame
print("Final DataFrame with One-Hot Encoding:")
display(df_final)
````

# Question 7: Consider the following dataset of exam scores: [50, 55, 52, 48, 51, 60, 62, 54, 53, 47, 65, 90] and detect the outlier using IQR method, and also plot the box-plot.

````python
import numpy as np
import matplotlib.pyplot as plt

# Dataset
scores = [50, 55, 52, 48, 51, 60, 62, 54, 53, 47, 65, 90]

# 1. Detect Outliers using IQR method
Q1 = np.percentile(scores, 25)
Q3 = np.percentile(scores, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in scores if x < lower_bound or x > upper_bound]

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Detected Outliers: {outliers}")

# 2. Plot the box-plot
plt.figure(figsize=(8, 5))
plt.boxplot(scores, vert=False, patch_artist=True, tick_labels=['Exam Scores'])
plt.title('Box Plot of Exam Scores')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
````
